```python
import pandas as pd
import numpy as np

## first, let's read in the raw data from an excel sheet
## for now, we'll analyze the first sheet of the document
try:
    foods_df = pd.read_excel("Alltables.xlsx", sheet_name=0, header=0, index=None)
    
except:
    print("the file does not exist yo!")

finally:
    print("Done executing!!, Happy Coding")
    
    

## nice, now we obviously have to drop a few rows, which I'll make columns later
foods_df = foods_df.iloc[3:20, :]

## let's remove unnecessary colummns
## list of unnecessary columns
unnecessary_columns = [f'Unnamed: {i}' for i in range (22, 43) if i != 23]
foods_df = foods_df.drop(unnecessary_columns, axis=1)

## let's not forget columns 1 and 2
foods_df = foods_df.drop(['Unnamed: 1', 'Unnamed: 2'], axis=1)


## now let's name our columns
## first part of the column
columns = ['Food Group'] 
years = [str(i) for i in range(1999, 2018)]

## second part
columns.extend(years)

## last part
columns.append('Annual Growth')

## finally
foods_df.columns = columns


## now the forth row isn't doing anything either so let's clear it
foods_df = foods_df.drop([4], axis=0)

## Data is looking clean now

## the last thing i'd like to do is remove 1/, /2 and /3 from rows 3,7 and 18 respectively, also, the first row
## isn't necessary
foods_df.iloc[0, 0] = 'Total U.S food Imports'

foods_df.iloc[3, 0] = 'Fish and Shellfish'

foods_df.iloc[14, 0] = 'Beverages'

foods_df = foods_df.drop([3], axis=0)


## gorgeous, now we can do any analysis we want to do
print("The end")


## let's now take a look at what our data looks like
foods_df

```
